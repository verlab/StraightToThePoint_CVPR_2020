```
! pip install youtube-downloader

import torch
import nltk
nltk.download('punkt')

model = torch.hub.load('washingtonsk8/StraightToThePoint_CVPR2020:main', 'SemanticFastForward_RL', pretrained=True, progress=True)
model.cuda()
model.eval()
```

```
VIDEO_ID = 'xkKuIlYSMMU'
document = ['Insert the skewer into the hot dogs',
            'Combine flour sugar polenta baking powder egg and milk',
            'Heat oil in a pan',
            'Pour the batter over the hot dog',
            'Deep fry the corn dog in oil']

! youtube-dl https://www.youtube.com/watch?v=$VIDEO_ID -o $VIDEO_ID --merge-output-format mp4 # Download Video File

sf = model.fast_forward_video(f'/content/{VIDEO_ID}.mp4', document)
```

```
import numpy as np
from matplotlib import pyplot as plt
plt.scatter(sf[:-1], np.array(sf[1:])-np.array(sf[:-1]))
plt.title('Agent Skips')
plt.xlabel('Frame Number')
plt.ylabel('Skip')
plt.show()
```
